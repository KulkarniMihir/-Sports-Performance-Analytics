## Part 1- Data Coding and Merging

```python
#Import Libraries

import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import NHL Team Data and Display
NHL_Team=pd.read_csv("Assignment Data/Week 2/NHL_team.csv")
display(NHL_Team)
```

```python
NHL_Team.drop(['Unnamed: 0','tname','lname','abbr','sname'], axis=1, inplace=True)
NHL_Team.head()
```

```python
NHL_Team.rename(columns={'name':'team_name'}, inplace=True)
NHL_Team.head()
```

```python
#Import NHL Competition Data and Display Head
NHL_Competition=pd.read_csv("Assignment Data/Week 2/NHL_competition.csv")
NHL_Competition.head()
```

```python
NHL_Competition.drop(['Unnamed: 0','tz','start','end',], axis=1, inplace=True)
NHL_Competition.rename(columns={'name':'competition_name'}, inplace=True)
NHL_Competition.head()
```

```python
#Import NHL Game Data and Display Head
NHL_Game=pd.read_csv("Assignment Data/Week 2/NHL_game.csv")
NHL_Game.head()
```

```python
NHL_Game.drop(['X','period','status',], axis=1, inplace=True)
NHL_Game.head()
```

```python
#Import NHL PPPK Data and Display Head
NHL_PPPK=pd.read_csv("Assignment Data/Week 2/pp.pk.ppgf.csv")
NHL_PPPK.head()
```

```python
NHL_Game=pd.merge(NHL_Team, NHL_Game, on=['tid'])
NHL_Game.head()
```

```python
NHL_Game=pd.merge(NHL_Competition, NHL_Game, on=['comp_id'])
NHL_Game.head()
```

```python
NHL_Game['hgd']=NHL_Game['hscore']-NHL_Game['ascore']
NHL_Game.head()
```

```python
NHL_Game.info()
```

```python
NHL_Game=NHL_Game[pd.notnull(NHL_Game["hgd"])]
NHL_Game.head()
```

```python
NHL_Game.isnull().sum()
```

```python
import datetime
NHL_Game['date']=pd.to_datetime(NHL_Game['date'])
NHL_Game.info()
```

```python
NHL_Game.head()
```

```python
NHL_Game.sort_values(by=['date'])
```

```python
NHL_Home=NHL_Game[NHL_Game['home_away'] == 'home']
NHL_Home.shape
```

```python
NHL_Away=NHL_Game[NHL_Game['home_away'] == 'away']
NHL_Away.shape
```

```python
NHL_Home.rename(columns={'ascore':'goals_against','hscore':'goals_for'}, inplace=True)
NHL_Away.rename(columns={'ascore':'goals_for','hscore':'goals_against'}, inplace=True)
```

```python
NHL_Home['win']=np.where(NHL_Home['goals_for']>NHL_Home['goals_against'],1,(np.where(NHL_Home['goals_for']<NHL_Home['goals_against'],0,0.5)))
NHL_Home.head()
```

```python
NHL_Away['win']=np.where(NHL_Away['goals_for']>NHL_Away['goals_against'],1,(np.where(NHL_Away['goals_for']<NHL_Away['goals_against'],0,0.5)))
NHL_Away.head()
```

```python
NHL_Game = NHL_Home
NHL_Game = NHL_Game.append(NHL_Away)
NHL_Game.shape
```

```python
NHL_Game.tail()
```

```python
NHL_Team_Stats = NHL_Game.groupby(['tid', 'competition_name', 'type'])['win','goals_for','goals_against'].sum().reset_index()
display(NHL_Team_Stats)
```

```python
NHL_Game_Count = NHL_Game.groupby(['tid', 'competition_name', 'type'])['win'].count().reset_index(name="game_count")
display(NHL_Game_Count)
```

```python
NHL_Team_Stats = pd.merge(NHL_Game_Count, NHL_Team_Stats, on=['tid', 'competition_name', 'type'])
```

```python
NHL_Team_Stats = pd.merge(NHL_Team_Stats, NHL_Team, on=['tid'])
NHL_Team_Stats.head()
```

```python
NHL_Team_Stats = pd.merge(NHL_Team_Stats, NHL_PPPK, on=['tricode', 'competition_name'])
NHL_Team_Stats.head()
```

```python
NHL_Team_Stats['win_pct']=NHL_Team_Stats['win']/NHL_Team_Stats['game_count']
NHL_Team_Stats['avg_gf']=NHL_Team_Stats['goals_for']/NHL_Team_Stats['game_count']
NHL_Team_Stats['avg_ga']=NHL_Team_Stats['goals_against']/NHL_Team_Stats['game_count']
NHL_Team_Stats.head()
```

```python
NHL_Team_R_Stats = NHL_Team_Stats[NHL_Team_Stats['type'] == 2]
display(NHL_Team_R_Stats)
```
